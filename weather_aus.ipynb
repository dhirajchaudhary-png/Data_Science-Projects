```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
df['Evaporation']=lb.fit_transform(df['Evaporation'])
df['Sunshine']=lb.fit_transform(df['Sunshine'])
df['WindGustDir']=lb.fit_transform(df['WindGustDir'])
df['WindGustSpeed']=lb.fit_transform(df['WindGustSpeed'])
df['WindDir9am']=lb.fit_transform(df['WindDir9am'])
df['WindDir3pm']=lb.fit_transform(df['WindDir3pm'])
df['MinTemp']=lb.fit_transform(df['MinTemp'])
df['MaxTemp']=lb.fit_transform(df['MaxTemp'])
df['WindSpeed9am']=lb.fit_transform(df['WindSpeed9am'])
df['WindSpeed3pm']=lb.fit_transform(df['WindSpeed3pm'])
df['Humidity9am']=lb.fit_transform(df['Humidity9am'])
df['Humidity3pm']=lb.fit_transform(df['Humidity3pm'])
df['Pressure9am']=lb.fit_transform(df['Pressure9am'])
df['Pressure3pm']=lb.fit_transform(df['Pressure3pm'])
df['Cloud9am']=lb.fit_transform(df['Cloud9am'])
df['Cloud3pm']=lb.fit_transform(df['Cloud3pm'])
df['Temp9am']=lb.fit_transform(df['Temp9am'])
df['Temp3pm']=lb.fit_transform(df['Temp3pm'])
df['RainToday']=lb.fit_transform(df['RainToday'])
df['RainTomorrow']=lb.fit_transform(df['RainTomorrow'])
```

```python
df['WindDir3pm']=df['WindDir3pm'].fillna('16')
df['WindDir9am']=df['WindDir9am'].fillna('5')
df['MinTemp']=df['MinTemp'].fillna('12.1842')
df['MaxTemp']=df['MaxTemp'].fillna('23.235')
df['Evaporation']=df['Evaporation'].fillna('5.4725155')
df['Sunshine']=df['Sunshine'].fillna('7.63053')
df['WindGustDir']=df['WindGustDir'].fillna('0')
df['WindGustSpeed']=df['WindGustSpeed'].fillna('39.97051')
df['WindSpeed9am']=df['WindSpeed9am'].fillna('14.0123')
df['WindSpeed3pm']=df['WindSpeed3pm'].fillna('18.63114')
df['Humidity9am']=df['Humidity9am'].fillna('68.826832')
df['Humidity3pm']=df['Humidity3pm'].fillna('51.4492')
df['Pressure9am']=df['Pressure9am'].fillna('1017.654577')
df['Pressure3pm']=df['Pressure3pm'].fillna('1015.2579')
df['Cloud9am']=df['Cloud9am'].fillna('7.0')
df['Cloud3pm']=df['Cloud3pm'].fillna('7.0')
df['Temp9am']=df['Temp9am'].fillna('16.987663')
df['Temp3pm']=df['Temp3pm'].fillna('21.6931826')
```

```python
df.info()
```

```python
continous_data=['MinTemp', 'MaxTemp', 'Rainfall', 'Evaporation',
       'Sunshine', 'WindGustSpeed',
       'WindSpeed9am', 'WindSpeed3pm', 'Humidity9am', 'Humidity3pm',
       'Pressure9am', 'Pressure3pm','Temp9am',
       'Temp3pm']
df_cleaned=df.copy()
def replace_outlier(df,column):
    Q1=df[column].quantile(0.25)
    Q3=df[column].quantile(0.75)

    IQR=Q3-Q1
    lower_bound=Q1-1.5*IQR
    upper_bound=Q3-1.5*IQR

    df[column]=df[column].apply(lambda x:Q1 if x<lower_bound else Q3 if x>upper_bound else x)
    return df
```

```python
for column in continous_data:
    df_cleaned=replace_outlier(df_cleaned,column)
```

```python
df_cleaned['RainToday']=lb.fit_transform(df_cleaned['RainToday'])
df_cleaned['RainTomorrow']=lb.fit_transform(df_cleaned['RainTomorrow'])
df_cleaned['Date']=lb.fit_transform(df_cleaned['Date'])
df_cleaned['Location']=lb.fit_transform(df_cleaned['Location'])
```

```python
x=df_cleaned[['MinTemp', 'MaxTemp', 'Rainfall', 'WindGustDir', 'WindGustSpeed', 'WindDir9am', 'WindDir3pm',
       'WindSpeed9am', 'WindSpeed3pm', 'Humidity9am', 'Humidity3pm','Cloud9am','Cloud3pm','Pressure9am', 'Pressure3pm', 'Temp9am',
       'Temp3pm', 'RainToday']]
y=df_cleaned['RainTomorrow']
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x=scaler.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=39)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression('liblinear')
lr=LogisticRegression()
lr.fit(x_train, y_train)
```

```python
predict1=lr.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
cm=confusion_matrix(y_test,predict1)
print(cm)

disp=ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=[0,1])
disp.plot(cmap='Blues')
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train,y_train)
```

```python
y_pred=knn.predict(x_test)
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
cm=confusion_matrix(y_test,y_pred)
disp=ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=[0,1])
disp.plot(cmap='Blues')
plt.show()
```
