```python
continous_columns=['radius_mean','texture_mean','perimeter_mean','area_mean','smoothness_mean','compactness_mean',
'concavity_mean',
'symmetry_mean',
'fractal_dimension_mean','smoothness_worst','concavity_se']
```

```python
from scipy.stats import zscore
for col in continous_columns:
    z_score=zscore(data[col])
    threshold=3
    data=data[(z_score<threshold)&(z_score>-threshold)]
```

```python
z_score
```

```python
continuos_data=[
       'concave points_mean', 'symmetry_mean', 
       'radius_se', 'texture_se', 'perimeter_se', 'area_se', 'smoothness_se',
       'compactness_se','concave points_se', 'symmetry_se',
       'fractal_dimension_se', 'radius_worst', 'texture_worst',
       'perimeter_worst', 'area_worst',
       'compactness_worst', 'concavity_worst', 'concave points_worst',
       'symmetry_worst', 'fractal_dimension_worst']
df_cleaned=data.copy()
def replace_outlier(data,column):
    Q1=data[column].quantile(0.25)
    Q3=data[column].quantile(0.75)

    IQR=Q3-Q1
    lower_bound=Q1-1.5*IQR
    upper_bound=Q3-1.5*IQR

    data[column]=data[column].apply(lambda x:Q1 if x<lower_bound else x)
    data[column]=data[column].apply(lambda x:Q3 if x>upper_bound else x)
    return data
```

```python
for columns in continuos_data:
    df_cleaned=replace_outlier(df_cleaned,columns)
```

```python
#used for huge dataset to transform float into integer
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
data['diagnosis']=lb.fit_transform(data['diagnosis'])
```

```python
# converting float into integer 
## best for categorical data
df_cleaned['diagnosis']=df_cleaned['diagnosis'].map({'B':0,'M':1})
df_cleaned['diagnosis']
```

```python
# step 1 dependent and independent variable
x=df_cleaned[['radius_mean', 'texture_mean', 'perimeter_mean',
       'area_mean', 'smoothness_mean', 'compactness_mean', 'concavity_mean',
       'concave points_mean', 'symmetry_mean', 'fractal_dimension_mean',
       'radius_se', 'texture_se', 'perimeter_se', 'area_se', 'smoothness_se',
       'compactness_se', 'concavity_se', 'concave points_se', 'symmetry_se',
       'fractal_dimension_se', 'radius_worst', 'texture_worst',
       'perimeter_worst', 'area_worst', 'smoothness_worst',
       'compactness_worst', 'concavity_worst', 'concave points_worst',
       'symmetry_worst', 'fractal_dimension_worst']]
y=df_cleaned['diagnosis']
```

```python
# step 2 train test split
from sklearn.model_selection import train_test_split
x_train, x_test,y_train, y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
# importing model
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train,y_train)
```

```python
# make prediction
y_pred=knn.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
acc=accuracy_score(y_test,y_pred)
```

```python
acc
```

```python
# for contionous datasets accuracy determines the quality of algo 
# for categorical dataset confusion matrix, precision score, recall score matters
```

```python
from sklearn.metrics import precision_score, f1_score, recall_score
ps=precision_score(y_test,y_pred)
f1=f1_score(y_test, y_pred)
rs=recall_score(y_test,y_pred)
```
