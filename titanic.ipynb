# methods to remove outlier
# we have to option to solve it 
# zscore, iQR
# zscore- if data is normally distributed
# IQR- if data is not normally distributed

# IQR =Inter Quartile Range

# Step 1
# first quartle(q1)= this is median of lower half of the data 
# third quartile (q3)= this is the media  of the upper half of the data

# step 2
# IQR=Q3-Q1

# Step 3
# lower bound= Q1-1.5*IQR
# Upper bound=Q3+1.5*IQR

# any datapoint above upperbound and lower bound called outlier





```python
numerical_col=['Age','Fare']
data = df[col].dropna() 
df_cleaned=df.copy()

def replace_outlier_iqr(df,column):
    # step 1
    q1=df[column].quantile(0.25)
    q3=df[column].quantile(0.75)


    # step 2- calculate iqr
    IQR=q3-q1

    # step 3- upper bound and lower bound
    lower_bound=q1-1.5*IQR
    upper_bound=q3+1.5*IQR

    # replace values with below the lower bound with q1
    df[column]=df[column].apply(lambda x:q1 if x<lower_bound else x)

    #replace values above the upper bound with q3
    df[column]=df[column].apply(lambda x:q3 if x>upper_bound else x)

    return df

for column in numerical_col:
    df_cleaned=replace_outlier_iqr(df_cleaned,column)
    
```
